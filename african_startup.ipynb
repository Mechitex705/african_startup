# Distribution of Funding Amount

```python
plt.figure(figsize=(12, 6))
sns.histplot(data=data, x="Amount (US$)", bins=30, kde=True)
plt.xlabel("Amount (in US$)")
plt.title("Distribution of Funding Amount")
plt.show()
```

### Funding Distribution
The above graph shows the distribution of funding amount for projects. The x-axis shows the amount of funding in US dollars, and the y-axis shows the number of projects.
The most notable thing about the distribution is the long tail. This means that there are a few projects that received a lot of funding, but there are also many projects that received very little funding.
The distribution also shows that the probability of a project being funded decreases as the funding amount increases. This is likely because it is more difficult to raise a large amount of funding.

## Funding by Country

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=data, x="Location")
plt.xticks(rotation=180)
plt.xlabel("Location")
plt.ylabel("Location")
plt.title("Tech Startups by Country")
plt.show()
```

#### Country Analysis 
The above bar graph shows the best startups by country in Africa. The countries are ranked by the number of startups they have.
The most notable thing about the graph is that Nigeria has the most startups, followed by Egypt and Kenya. This suggests that these countries are the most conducive to startup growth in Africa.
The graph also shows that there is a growing number of startups in Africa. This is likely due to a number of factors, such as the increasing availability of funding, the growing pool of talent, and the improving business environment.

# Funding Status

```python
plt.figure(figsize=(11, 6))
sns.countplot(data=data, x="Debt/Equity")
plt.xlabel("Debt/Equity")
plt.ylabel("Count")
plt.title("Funding Status of Tech Startups")
plt.show()
```

### Funding Status
The image shows the funding status of tech startups by type of funding. The most common type of funding is equity funding, followed by debt funding and grants.
Here are some insights that can be drawn from the data:

    Equity funding is the most common type of funding for tech startups. This is because equity funding provides startups with the capital they need to grow their business without having to repay the loan.
    Debt funding is the second most common type of funding for tech startups. This is because debt funding can be used to finance specific projects or expenses.
    Grants are the least common type of funding for tech startups. This is because grants are typically awarded to startups that are working on innovative or socially beneficial projects.

# Correlation heatmap

```python
correlation_matrix = data.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

### Correlation Analysis 
The correlation heatmap shows the correlation between the number of employees, the amount of money spent, and whether or not a new product was launched. The darker the cell, the stronger the correlation.

The most interesting thing to note is the strong correlation between the number of employees and whether or not a new product was launched (0.64). This means that the more employees a company has, the more likely it is to launch a new product. This is likely because larger companies have more resources and manpower to devote to product development.

There is also a weak negative correlation between the amount of money spent and whether or not a new product was launched (-0.2). This means that spending more money does not necessarily guarantee that a new product will be launched. However, it is important to note that the correlation is weak, so there may be other factors at play.

Overall, the correlation heatmap suggests that the number of employees is the most important factor in determining whether or not a new product will be launched. However, the amount of money spent also plays a role, albeit to a lesser extent.
